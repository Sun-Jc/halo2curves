```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("benchmark.csv")
```

```python
df
```

```python
df['small_ratio'] = df['NumSmallScalar'] / df['BatchSize']
df['MSMTime(ns)'] = df['MSMTime(ns)'] / 1e9
```

```python
df
```

```python
df1 = df[df['small_ratio'] < 0.3]
df_grouped = df1.groupby('BatchSize')['MSMTime(ns)'].mean()
plt.plot(df_grouped, marker='o', linestyle='-',  label="random")

df1 = df[df['small_ratio'] < 0.96]
df1 = df1[df1['small_ratio'] >= 0.3]
df_grouped = df1.groupby('BatchSize')['MSMTime(ns)'].mean()
plt.plot(df_grouped, marker='*', linestyle='-', label="mixed")

df1 = df[df['small_ratio'] >= 0.96]
df1 = df1[df1['ApproachId']==0]
df_grouped = df1.groupby('BatchSize')['MSMTime(ns)'].mean()
plt.plot(df_grouped, marker='x', linestyle='-', label="small")

# df1 = df[df['small_ratio'] >= 0.96]
# df1 = df1[df1['ApproachId']==1]
# df_grouped = df1.groupby('BatchSize')['MSMTime(ns)'].mean()
# plt.plot(df_grouped, marker='x', linestyle='-', label="small(modified, not fully optimized)")


# Labels and title
plt.xlabel('Batch Size')
plt.ylabel('Time (s)')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
